```python
usuarios = {}

def registrar_usuario():
  usuario = input("Ingrese nombre de usuario: ").strip()
  if usuario in usuarios:
    print("¡Error! El usuario ya existe.")
    return False
  contrasena = input("Ingrese contraseña: ").strip()
  usuarios[usuario] = contrasena
  print("¡Usuario registrado exitosamente!")
  return True

def login_usuario():
  usuario = input("Ingrese usuario: ").strip()
  if usuario not in usuarios:
        print("¡Error! Usuario no encontrado.")
        return False
  contrasena = input("Ingrese contraseña: ").strip()
  if usuarios[usuario] == contrasena:
        print(f"¡Login exitoso para {usuario}!")
        return True
  else:
        print("¡Contraseña incorrecta!")
        return False

def mostrar_usuarios():
  if not usuarios:
      print("No hay usuarios registrados.")
      return
  print("\n--- Lista de Usuarios ---")
  for usuario in usuarios:
        print(f"Usuario: {usuario} (registrado)")
  print("------------------------")

while True:
    print("\n=== MENÚ PRINCIPAL ===")
    print("1. Registrar usuario")
    print("2. Loguear usuario")
    print("3. Mostrar usuarios")
    print("4. Salir")

    opcion = input("Elija una opción (1-4): ").strip()

    if opcion == "1":
        registrar_usuario()
    elif opcion == "2":
        login_usuario()
    elif opcion == "3":
        mostrar_usuarios()
    elif opcion == "4":
        print("¡Gracias por usar el sistema!")
        break
    else:
        print("Opción inválida. Intente nuevamente.")

```
